## Import Pandas library

```python
import pandas as pd
```

## Import dataframe

```python
df_orig = pd.read_excel('Dataset.xlsx')
df_orig.head(5)
```

## Remove unnecessary columns

```python
df2 = df_orig.drop(columns=['Unnamed: 0', 'X', 'ID'])
df2.head(5)
```

## Convert original dataset to 1st Normal Form (1NF)

```python
#Convert Directors, Genres, Country, and Language columns to explodable format using split function
df2 = df2.assign(Directors=df2.Directors.str.split(","))
df2 = df2.assign(Genres=df2.Genres.str.split(","))
df2 = df2.assign(Country=df2.Country.str.split(","))
df2 = df2.assign(Language=df2.Language.str.split(","))
df2.head(5)
```

```python
#Explode dataframe
col_to_explode = ['Directors','Genres','Country','Language']

for i in col_to_explode:
    df2 = df2.explode(i)

#reset index
df2 = df2.reset_index()

#delete old index
df2 = df2.drop(['index'],axis=1)
```
